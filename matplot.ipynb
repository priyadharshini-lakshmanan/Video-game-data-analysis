```python
# Concatenate along columns
df3 = pd.concat([df1, df2], axis=1)

# Save the merged DataFrame to a new CSV
df3.to_csv('merged_side_by_side.csv', index=False)
```

```python
df3
```

```python
import matplotlib.pyplot as plt

# Group and sort top 10 genres by global sales
genre_sales = df3.groupby("Genre")["Global_Sales"].sum().sort_values(ascending=False).head(10)

# Create unique colors for each bar
colors = plt.cm.tab10.colors  # 10 visually distinct colors

# Plot manually using plt.bar
plt.figure(figsize=(10, 6))
bars = plt.bar(genre_sales.index, genre_sales.values, color=colors)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.1f}', ha='center', va='bottom', fontsize=9)

# Add title and labels
plt.title("Top 10 Game Genres by Global Sales", fontsize=14)
plt.ylabel("Global Sales (in millions)")
plt.xlabel("Genre")
plt.xticks(rotation=45)

# Add legend
plt.legend(bars, genre_sales.index, title="Genres", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Select the column to group by (e.g., 'Genre')
group_col = 'Genre'

plt.figure(figsize=(10, 7))

# Loop through each genre and plot separately
for genre, group_data in df3.groupby(group_col):
    plt.scatter(group_data["rating"], group_data["Global_Sales"],
                label=genre, alpha=0.6, s=50)

plt.title(" Rating vs Global Sales by Genre")
plt.xlabel("Rating")
plt.ylabel("Global Sales")
plt.grid(True)

# Add legend outside the plot
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#3.platforms have the most games with high ratings (e.g., above 4)
import matplotlib.pyplot as plt

# Filter high-rated games
high_rated = df3[df3["rating"] > 4]

# Count top platforms with high-rated games
platform_counts = high_rated["Platform"].value_counts().head(10)

# Get 10 distinct colors
colors = plt.cm.tab10.colors

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(platform_counts.index, platform_counts.values, color=colors)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             f'{int(height)}', ha='center', va='bottom', fontsize=9)

# Add chart title and labels
plt.title("Top Platforms with High-Rated Games (Rating > 4)", fontsize=14)
plt.ylabel("Number of Games")
plt.xlabel("Platform")
plt.xticks(rotation=45)

# Add legend manually
plt.legend(bars, platform_counts.index, title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract release year
df3["release_year"] = pd.to_datetime(df3["release_date"], errors='coerce').dt.year

# Group by year and sum global sales
trend = df3.groupby("release_year")[["Global_Sales"]].sum().dropna()
trend = trend.sort_index()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(trend.index, trend["Global_Sales"], color='tomato', marker='o', linewidth=2)

# Add value labels to each point
for x, y in zip(trend.index, trend["Global_Sales"]):
    plt.text(x, y + 0.3, f"{y:.1f}", ha='center', va='bottom', fontsize=8)

# Show all years as integers
years = trend.index.astype(int)
plt.xticks(ticks=trend.index, labels=years, rotation=45)

# Titles and formatting
plt.title("Global Sales and releases trend Over Time", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Global Sales (in millions)")

# ✅ Remove grid
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
combo = df3.groupby(["Genre", "Platform"])["Global_Sales"].sum().sort_values(ascending=False).head(15)
combo_df = combo.unstack().fillna(0)
# Plot
ax = combo_df.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title("Top performing Genre + Platform Combinations by Sales")
plt.ylabel("Global Sales")
plt.xlabel("Genre")
#value labels
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{height:.1f}',
                ha='center', va='center', fontsize=8, color='white'
            )

plt.tight_layout()
plt.show()
```
