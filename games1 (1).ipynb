```python
def convert_k_values(value):
    """
    Convert values with 'K' suffix to numeric values by multiplying by 1000.
    
    Args:
        value: Input value (string, int, or float)
        
    Returns:
        Numeric value with 'K' converted to thousands
    """
    if pd.isna(value):
        return value
    
    # Convert to string for processing
    str_value = str(value).strip()
    
    # Check if the value ends with 'K' (case insensitive)
    if str_value.upper().endswith('K'):
        # Extract the numeric part before 'K'
        numeric_part = str_value[:-1]
        try:
            # Convert to float and multiply by 1000
            return float(numeric_part) * 1000
        except ValueError:
            # If conversion fails, return original value
            return value
    else:
        # Try to convert to numeric, return as is if already numeric
        try:
            return float(str_value)
        except ValueError:
            # If it's not numeric, return original value
            return value

```

```python
def process_dataframe_columns(df, columns=None):
    """
    Apply the K conversion to specified columns in a DataFrame.
    
    Args:
        df: pandas DataFrame
        columns: list of column names to process (if None, processes all columns)
        
    Returns:
        DataFrame with converted values
    """
    df_copy = df.copy()
    
    if columns is None:
        columns = df_copy.columns
    
    for col in columns:
        if col in df_copy.columns:
            df_copy[col] = df_copy[col].apply(convert_k_values)
    
    return df_copy
```
