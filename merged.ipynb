```python
# Concatenate along columns
df3 = pd.concat([df1, df2], axis=1)

# Save the merged DataFrame to a new CSV
df3.to_csv('merged_side_by_side.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Group and sort top 10 genres by global sales
genre_sales = df3.groupby("Genre")["Global_Sales"].sum().sort_values(ascending=False).head(10)

# Create unique colors for each bar
colors = plt.cm.tab10.colors  # 10 visually distinct colors

# Plot manually using plt.bar
plt.figure(figsize=(10, 6))
bars = plt.bar(genre_sales.index, genre_sales.values, color=colors)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.1f}', ha='center', va='bottom', fontsize=9)

# Add title and labels
plt.title("Top 10 Game Genres by Global Sales", fontsize=14)
plt.ylabel("Global Sales (in millions)")
plt.xlabel("Genre")
plt.xticks(rotation=45)

# Add legend
plt.legend(bars, genre_sales.index, title="Genres", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Key Observation:
* Action games lead with 1751.2 million in global sales, making them the most commercially successful genre.
* Sports and Shooter genres follow, with 1330.9 million and 1037.4 million in sales respectively.
* Other strong performers include Role-Playing, Platform, and Miscellaneous genres.
* Genres like Fighting, Simulation, and Puzzle generate significantly less sales in comparison.

Insights:
* Action and Sports games dominate the market, indicating strong player interest and commercial viability.
* Developers targeting high-revenue genres should focus on Action, Sports, and Shooter titles.
* Lesser-performing genres like Puzzle or Simulation may cater to niche audiences or need innovation to grow.

```python
# Select the column to group by (e.g., 'Genre')
group_col = 'Genre'

plt.figure(figsize=(8, 6))

# Loop through each genre and plot separately
for genre, group_data in df3.groupby(group_col):
    plt.scatter(group_data["rating"], group_data["Global_Sales"],
                label=genre, alpha=0.6, s=50)

plt.title(" Rating vs Global Sales by Genre")
plt.xlabel("Rating")
plt.ylabel("Global Sales")
plt.grid(True)

# Add legend outside the plot
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Key Observation:
* Most games are rated between 3.0 and 4.5, showing a concentration of user ratings in that range.
* Global sales are mostly below 10 million, with only a few games exceeding that, indicating a small number of blockbuster titles.
* The highest-selling games do not always have the highest ratings—some top sellers have moderate ratings around 3.5 – 4.0.
* Genres like Action, Sports, and Racing dominate the higher sales range, regardless of their user rating.

Insights:
* Some highly rated games sell well, but others don’t, so rating alone doesn’t decide a game’s success.
* Genres like Action and Sports often sell better even if they have average ratings, showing that popularity or fan base matters too.
* The dots are spread out, so we can't say that higher ratings always lead to higher sales.
* A small number of games have both high ratings and high sales, making them standout successes.


```python
#3.platforms have the most games with high ratings (e.g., above 4)
import matplotlib.pyplot as plt

# Filter high-rated games
high_rated = df3[df3["rating"] > 4]

# Count top platforms with high-rated games
platform_counts = high_rated["Platform"].value_counts().head(10)

# Get 10 distinct colors
colors = plt.cm.tab10.colors

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(platform_counts.index, platform_counts.values, color=colors)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             f'{int(height)}', ha='center', va='bottom', fontsize=9)

# Add chart title and labels
plt.title("Top Platforms with High-Rated Games (Rating > 4)", fontsize=14)
plt.ylabel("Number of Games")
plt.xlabel("Platform")
plt.xticks(rotation=45)

# Add legend manually
plt.legend(bars, platform_counts.index, title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Key Observation:
* This bar chart shows which gaming platforms have the most games rated above 4. The platforms are listed on the x-axis,

* PS2 (PlayStation 2) leads the list with 81 high-rated games, making it the most consistent in quality.
* It's followed by PS3 (63), Xbox 360 (48), and PlayStation (PS1) (45).
* Platforms like PSP, PC, and Game Boy (GB) have fewer high-rated titles, each with around 13–15.

Insights:
* Sony's PlayStation platforms dominate, with PS2,PS3 appearing in the top 10—showing strong quality across generations.
* Older platforms like PS2 and PS1 have more highly rated games than newer ones like PS4, indicating long-term developer support and broader libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract release year
df3["release_year"] = pd.to_datetime(df3["release_date"], errors='coerce').dt.year

# Group by year and sum global sales
trend = df3.groupby("release_year")[["Global_Sales"]].sum().dropna()
trend = trend.sort_index()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(trend.index, trend["Global_Sales"], color='tomato', marker='o', linewidth=2)

# Add value labels to each point
for x, y in zip(trend.index, trend["Global_Sales"]):
    plt.text(x, y + 0.3, f"{y:.1f}", ha='center', va='bottom', fontsize=8)

# Show all years as integers
years = trend.index.astype(int)
plt.xticks(ticks=trend.index, labels=years, rotation=45)

# Titles and formatting
plt.title("Global Sales and releases trend Over Time", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Global Sales (in millions)")

#Remove grid
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
Key Observation:
* From 1980 to around 2007, global game sales slowly but steadily increased.
* Sales started to grow much faster from 2008 onwards, showing high demand for games.
* The highest sales were recorded in 2022, with 569.2 million units sold.
* After 2022, there is a steep decline in global sales, especially in 2023 and 2024.

Insights:
* The video game industry has grown massively over the decades, especially after 2005.
* 2022 was a record-breaking year, possibly due to pandemic-related gaming habits or major releases.
* The sudden drop after 2022 might reflect data cutoff, market saturation, or delayed releases in recent years.
```
